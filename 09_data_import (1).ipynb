Import the ðŸ¦¸ðŸ¾â€â™€ï¸"Avengers"ðŸ¦¹ðŸ» data from the [FiveThirtyEight data repository on GitHub](https://github.com/fivethirtyeight/data/blob/master/avengers/avengers.csv).

```r
avengers_data = read_csv(
    'https://raw.githubusercontent.com/fivethirtyeight/data/master/avengers/avengers.csv',
    col_types = cols()
)
```

```r
avengers_data |> head()
```

Make an ECDF plot ðŸ“Š of `Appearances`.

(â—ï¸Hint: combine `arrange`, `mutate` and `row_number`/`rank` to produce a new column to encode with y-position.)

```r
avengers_data |>
    select(`Name/Alias`, Appearances) |>
    arrange(Appearances) |> 
    mutate(rk = row_number() / n())
    
```

```r
avengers2 = avengers_data |> 
    mutate(ypos = rank(Appearances, ties.method = 'first') / n()) |>
    select(`Name/Alias`, Appearances, ypos) |>
    arrange(desc(ypos))
```

```r
avengers3 = avengers_data |>
    select(`Name/Alias`, Appearances) |>
    arrange(Appearances) |>
    mutate(ypos = row_number() / n())
```

```r
avengers3 |>
    filter (ypos > 0.80, ypos< 0.85)
```

```r
highlight = avengers3 |> filter (`Name/Alias` == 'James Buchanan Barnes')
```

```r
p = ggplot(avengers3, aes(x = Appearances, y = ypos)) +
    geom_point() +
    geom_line() +
    geom_text_repel(
        data = highlight, 
        mapping = aes(label = `Name/Alias`),
        hjust = 0,
        nudge_x = 0.5,
        nudge_y = -0.15,
        box.padding = 0.5,
        segment.curvature = 1,
        segment.ncp = 2,
        segment.angle = 0
    ) +
    scale_y_continuous(labels = scales::percent_format())

p
```
