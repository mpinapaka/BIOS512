Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
p = ggplot(penguins, aes(x = body_mass_g))+
    geom_histogram(fill = 'steel blue')+ 
    facet_wrap(~year, ncol=1, scale = 'free_y') + 
    theme_minimal() +
    theme(strip.text = element_text(hjust = 0.9), plot.title.position = 'panel') +
    labs(x = "Body Mass in grams", y = "Number of Penguins", title='Histogram of penguin body mass for measurements
    taken from 2007-2009') + 
    scale_x_continuous(labels = comma_format())
    
    

p
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
options(repr.plot.width=4, repr.plot.height=4, repr.plot.res = 300)
p = ggplot(penguins, aes(x = sex, y = body_mass_g))+ 
    geom_point(na.rm = T , position = position_jitter(h = 0, w = 0.3), size = 0.8, alpha = 0.5)+ 
    labs(y = "Body Mass in grams", x = '')+
    coord_flip()+
    theme_minimal() +
    facet_wrap(~species, ncol=1, scales = 'free_y')
p
```
