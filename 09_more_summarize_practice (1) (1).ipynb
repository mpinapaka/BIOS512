Use `filter` and `summarize` to produce two single row tables. One that has the median `temp` for January, and one that has the median `temp` for June.

```r
# E.g.

median_january  = weather |> 
    group_by(month) |>
    filter(month == 1) |> 
    summarize(median.temp = median(temp))

jan_weather |> head()
```

Re-create the following figure.

<img src="./example.png" style="width: 800px;" />

- Use `geom_hline` for the median line.
- Use `scale_color_viridis_c` to set the color scale.
- This chart can be made by facetting or each plot can be created by itself and combined with `patchwork`. For this example use `patchwork`.
- Use `ylim` to make the y-axis for each plot the same. 

```r
#Two ways to make the graph: Can use functions or just make both graphs by hand 
```

```r
make_month_plot = function( month_to_plot ){
    month_weather = weather |> filter(month == month_to_plot)

    median_for_month  = weather |> 
        group_by(month) |>
        filter(month == month_to_plot) |> 
        summarize(median.temp = median(temp))

    p = ggplot(month_weather, aes(x = day, y = temp, color = hour)) + 
        geom_point(position = position_jitter(height = 0, width = 0.3), size = 1) +  
        scale_color_viridis_c() + 
        geom_hline(data = median_for_month, mapping = aes(yintercept = median.temp))
p
}
```

```r
make_month_plot (1) + make_month_plot (6)
```

```r
january = ggplot(jan_weather, aes(x = day, y = temp, color = hour)) + 
    geom_point(alpha= 0.5, na.rm = TRUE,position = position_jitter(h = 0, w = 0.2)) + 
    scale_color_viridis_c() +
    geom_hline (data = median_january,mapping = aes(yintercept = 37.04))
january
```

```r
june = ggplot(june_weather, aes(x = day, y = temp, color = hour)) + 
    geom_point(alpha= 0.5, na.rm = TRUE,position = position_jitter(h = 0, w = 0.2)) + 
    scale_color_viridis_c() +
    geom_hline (data = median_june,mapping = aes(yintercept = 71.96))
june
```

```r
options(repr.plot.width = 10, repr.plot.height = 4)

january = january + ylim(c(0, 90))

june = june + ylim(c(0, 90))

january + june
```
