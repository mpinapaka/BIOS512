```r
p = ggplot(penguins, aes(x = bill_length_mm)) + geom_histogram(na.rm = TRUE, binwidth= 1)

p
```

```r
p = ggplot(penguins, aes(x = bill_length_mm, y = species)) + geom_point(na.rm = TRUE, binwidth= 1)
p
```

```r
options(repr.plot.height = 3, repr.plot.width = 4)

p = ggplot(penguins, aes(x = bill_length_mm, y = species)) +
    geom_boxplot(na.rm = TRUE, outlier.alpha = 0) + 
    geom_point(na.rm = TRUE, position = position_jitter(height = 0.2, width = 0), shape = 21, color='firebrick')

p
```

```r
options(repr.plot.height = 3, repr.plot.width = 4)

p = ggplot(penguins, aes(x = bill_length_mm, fill=species, color = species)) +
    geom_histogram(na.rm=TRUE, binwidth = 1, position = 'identity', alpha = 0.5) +
    geom_rug()
p
```
